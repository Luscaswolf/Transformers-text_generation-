```python
import transformers
from transformers import pipeline
```

```python
gerador = pipeline('text-generation', model='pierreguillou/gpt2-small-portuguese')
```

```python
texto = "A Arte da Guerra (em chinês: 孫子兵法, transl. Método Militar de Sun Tzu) é um antigo tratado militar chinês que data do fim do período das Primaveras e Outonos"
resultado = gerador(texto, max_length=100, do_sample=True)
print(resultado)
```
